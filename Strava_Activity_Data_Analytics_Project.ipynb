```python
#Line Plot of Relative effort vs. Moving Time
import matplotlib.pyplot as plt

# Sort the DataFrame by 'Relative Effort'
df = df.sort_values('Relative Effort')

# Create a line plot
plt.figure(figsize=(10, 6))
plt.plot(df['Relative Effort'], df['Moving Time'], marker='o', linestyle='-')
plt.title('Line Plot of Relative Effort vs. Moving Time')
plt.xlabel('Relative Effort')
plt.ylabel('Moving Time')
plt.grid(True)
plt.show()
```

```python
# Box plot of your run distances by the day of the week
# Convert the 'Activity Date' column to datetime if it's not already
df['Activity Date'] = pd.to_datetime(df['Activity Date'])

# Create a 'Day of Week' column
df['Day of Week'] = df['Activity Date'].dt.day_name()

# Define the order of days for plotting
order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Create the box plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Day of Week', y='Distance', order=order)
plt.xticks(rotation=45)
plt.xlabel('Day of Week')
plt.ylabel('Distance (km)')
plt.title('Run Distances by Day of Week')
plt.tight_layout()
plt.show()
```

```python
# Create a heatmap of correlations
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True)
plt.show()
```

```python
import numpy as np
from sklearn.cluster import KMeans

# Select only the numeric columns from the DataFrame
numeric_columns = df.select_dtypes(include=np.number)

# Instantiate the KMeans model with the desired number of clusters
n_clusters = 3  # Adjust this based on the desired number of clusters
kmeans = KMeans(n_clusters=n_clusters)

# Fit the model to your running metrics data
kmeans.fit(df)

# Get the predicted cluster labels for each data point
labels = kmeans.labels_

# Get the cluster centers
cluster_centers = kmeans.cluster_centers_

# Visualize the clusters (for 2D data)
if numeric_columns.shape[1] == 2:
    plt.scatter(numeric_columns.iloc[:, 0], numeric_columns.iloc[:, 1], c=labels)
    plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], marker='x', color='red')
    plt.xlabel('Elapsed Time')
    plt.ylabel('Distance')
    plt.title('Running Metrics Clustering')
    plt.show()
```
